# Business Understanding


The objective of this analysis is to understand what drives medical insurance costs and quantify the impact of demographic and lifestyle factors. The project applies data analysis and predictive modeling techniques to uncover insights that can support better decision-making.</br>
### Stakeholders 
- **Insurance Companies**
    - Need to understand risk factors that drive higher medical cost
    - Aim to imrove risk assessment models
- **Individuals** 
  - Want transparency on what affects their insurance costs
  - Can make informed lifestyle decisions (e.g., smoking, BMI)

- **Data Sciencists**
  - Use this analysis as a baseline for building predictive pricing models
### Key Questions
This project aims to answer the following questions: </br>

1. What are the most important factors influencing medical insurance costs?
2. How significantly does smoking impact insurance charges?
3. What is the relationship between BMI and medical expenses?
4. Do individuals with dependents (children) incur higher costs?
5. Are there regional differences in insurance charges?
